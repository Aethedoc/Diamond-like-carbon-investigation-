```python
import numpy as np
import scipy as sci
from scipy.fftpack import fft
from matplotlib.axis import Axis
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
import pandas as pd
from scipy.interpolate import UnivariateSpline
import sys
np.set_printoptions(threshold=sys.maxsize) #Shows all values in array.
```

```python
def RamanDoped():
    
    Data = pd.read_csv("Raman fixed.csv") #.csv file input, all y-data should be not-offset.
    Data_X  =  Data['XAxis'] #X-axis are all the same,"X-Axis"
    Data_Seed= Data['Seeded']
    Data_T3 =  Data['BT3']
    Data_T15 = Data['BT15'] 
    Data_T30 = Data['BT30']
    Data_T45 = Data['BT45']
    Data_T60 = Data['BT60']
    Data_T75 = Data['BT75']
    #Data_T75S= Data['T75 Si'] #T75 Si investigation
    #Data_T75D= Data['T75 BULK'] #T75 Bulk investigation

    Data_XF = Data_X = Data_X.astype(float) #Setting String Dataframe -> float.
    Data_SeedF = Data_Seed = Data_Seed.astype(float) 
    Data_T3F = Data_T3 = Data_T3.astype(float) 
    Data_T15F = Data_T15 = Data_T15.astype(float) 
    Data_T30F = Data_T30 = Data_T30.astype(float) 
    Data_T45F = Data_T45 = Data_T45.astype(float)
    Data_T60F = Data_T60 = Data_T60.astype(float)
    Data_T75F = Data_T75 = Data_T75.astype(float)
    #Data_T75SF = Data_T75S = Data_T75S.astype(float)
    #Data_T75DF = Data_T75D = Data_T75D.astype(float)
    
    Data_list_X = Data_XF.values.tolist() #Setting float dataframe to list.
    Data_list_Seed = Data_SeedF.values.tolist()
    Data_list_T3 = Data_T3F.values.tolist()
    Data_list_T15 = Data_T15F.values.tolist()
    Data_list_T30 = Data_T30F.values.tolist()
    Data_list_T45 = Data_T45F.values.tolist()
    Data_list_T60 = Data_T60F.values.tolist()
    Data_list_T75 = Data_T75F.values.tolist()
    #Data_list_T75S = Data_T75SF.values.tolist()
    #Data_list_T75D = Data_T75DF.values.tolist()
    
    normT3  = [float(i)/max(Data_list_T3) for i in Data_list_T3] #normalizing raman data to maxiumum value (diamond peak)
    normT15 = [float(i)/max(Data_list_T15) for i in Data_list_T15]
    normT30 = [float(i)/max(Data_list_T30) for i in Data_list_T30]
    normT45 = [float(i)/max(Data_list_T45) for i in Data_list_T45]
    normT60 = [float(i)/max(Data_list_T60) for i in Data_list_T60]
    normT75 = [float(i)/max(Data_list_T75) for i in Data_list_T75]
    
    #print(normT3)
    def Graphing(): #Graphing function for normalized raman
        plt.title("Raman inspection of Boron-Doped Samples")
        plt.xlabel("Wavelength (cm^-1)")
        plt.ylabel("Intensity")
        #plt.ylim(200000, 500000) 
        #plt.xlim(1300, 1400)
        plt.grid(True)
        plt.rcParams["figure.figsize"] = 20,12
        
        x = Data_list_X #X-Axis
        
        ySeed = Data_list_Seed
        #plt.plot(x,ySeed)
        yT3 = normT3
        plt.plot(x, yT3)
        yT15 = normT15
        plt.plot(x, yT15)
        yT30 = normT30
        plt.plot(x, yT30)
        yT45 = normT45
        plt.plot(x, yT45)
        yT60 = normT60
        plt.plot(x, yT60)
        yT75 = normT75
        plt.plot(x, yT75)
        #yT75S = Data_list_T75S
        #plt.plot(x, yT75S)
        #yT75D = Data_list_T75D
        #plt.plot(x, yT75D)
        
        plt.legend(["T3", "T15", "T30", "T45", "T60", "T75"], loc ="upper left")
    Graphing() 
    
    def FWHM(): #function for Full width at half maximum
     
        Data_Xinc  =  Data['Xaxisinc'] #reversed x-axis for spline"
        Data_XincF = Data_Xinc = Data_Xinc.astype(float) #Setting String Dataframe -> float.
        Data_list_Xinc = Data_XincF.values.tolist() #Setting float dataframe to list.
        x = Data_list_Xinc[280:292] #280:292 is rough diamond peak for my data 
        blue = normT60[280:292] #diamond peak

        print(blue)
        
        import pylab as pl
        pl.plot(x,blue)    
        
        # creating a spline of x and blue-np.max(blue)/2 
        spline = UnivariateSpline(x, blue-np.max(blue)/2, s=0) 
        r1, r2 = spline.roots() # find the roots

        
        
        pl.plot(x, blue)
        pl.axvspan(r1, r2, facecolor='g', alpha=0.5)
        #pl.xlim(1310,1340)
        pl.show()
        print("Roots: ",r1,r2)
    #FWHM()
    
    
    
RamanDoped()
```
