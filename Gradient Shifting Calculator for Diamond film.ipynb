```python
import numpy as np
import scipy as sci
from scipy.fftpack import fft
from matplotlib.axis import Axis
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
import pandas as pd
import sys
np.set_printoptions(threshold=sys.maxsize) #Shows all values in array.
```

```python
#2 Variable: Si, Film and/or ETB CALC. [2 INPUTS ONLY]

def Gradient_2_Var():
    
    Data = pd.read_csv("BT15S1 GradientData.csv") #.csv inputs
    Data_X1 = Data['X1'] #Si
    Data_Y1 = Data['Y1']
    Data_X2 = Data['X2'] #Film
    Data_Y2 = Data['Y2']

    Data_X1F = Data_X1 = Data_X1.astype(float) #Setting String Dataframe -> float.
    Data_Y1F = Data_Y1 = Data_Y1.astype(float) 
    Data_X2F = Data_X2 = Data_X2.astype(float) 
    Data_Y2F = Data_Y2 = Data_Y2.astype(float) 
    
    Data_list_X1 = Data_X1F.values.tolist() #Setting float dataframe to list.
    Data_list_Y1 = Data_Y1F.values.tolist()
    Data_list_X2 = Data_X2F.values.tolist()
    Data_list_Y2 = Data_Y2F.values.tolist()
    #Calulating the Least square fit data line from raw data lists for a LOBF.
    #If data was in separate files, wouldnt have to specify legnth. range(len()) doesnt work. 
    Gradient1 = np.polyfit(Data_list_X1[0:231] , Data_list_Y1[0:231] , 1) #Finds [gradient, Y-intercept] using least squares fit via 1st degree poly
    print("Gradient, Si:", Gradient1) #Issue above, python thinks all lists = len(longest list) therefore "nan" for all blanks 
    Gradient2 = np.polyfit(Data_list_X2[0:395] , Data_list_Y2[0:395] , 1) #Finds [gradient, Y-intercept]
    print("Gradient, Film:", Gradient2)
    
    #Difference in gradients between "Background" Data and interested inclined "Growth" data. 
    Gradient_Film_Si = Gradient2-abs(Gradient1) #Diff in gradients between Si | film and Edge to Bulk (ETB)
    print("Gradient Difference:", Gradient_Film_Si)
    #Generating Lines Of Best Fit (LOBF) from above LINEST (Least squares fit) gradient.
    Film_Si_Y_DATA = []
    for i in range(len(Data_list_X2)): #m*x 
        Film_Si_Y = Gradient_Film_Si[0] * Data_list_X2[i]
        Film_Si_Y_DATA.append(Film_Si_Y)
    #print(ETB_Si_Y_DATA)
    
    #Generating Final Gradient Shifted Data (GSD) with sloped removed. 
    GSD_Film_Si = [] #GSD = Gradient shifted Data 
    for i in range(len(Data_list_Y2)):
        Film_Si_DIFF = Data_list_Y2[i] - abs(Film_Si_Y_DATA[i])
        GSD_Film_Si.append(Film_Si_DIFF)
    #print(GSD_Film_Si)
    
    def Graphing():
        plt.title("Gradient Shifted Film Edge by comparison of Si gradient | BT15")
        plt.xlabel("Length (µm)")
        plt.ylabel("Height (nm)")
        #plt.ylim(0, 30000) 
        #plt.xlim(3450, 3700)
        plt.grid(True)
        plt.rcParams["figure.figsize"] = 10,6
        
        xD1 = Data_list_X2 #Film RAW
        yD1 = Data_list_Y2
        plt.plot(xD1, yD1)
        xD2 = Data_list_X2 #Film-Si
        yD2 = GSD_Film_Si
        plt.plot(xD2, yD2)
        plt.legend(["Film RAW", "Film-Si"], loc ="lower left")
    Graphing() 
Gradient_2_Var()

```

```python
#3 Variable: Si, Film and ETB CALC. [REQUIRE ALL 3 INPUTS TO WORK]
def Gradient_3_Var():
    
    Data = pd.read_csv("BT15S1 GradientData.csv")
    Data_X1 = Data['X1'] #Si
    Data_Y1 = Data['Y1']
    Data_X2 = Data['X3'] #Film
    Data_Y2 = Data['Y3']
    Data_X3 = Data['X2'] #Edge-Bulk
    Data_Y3 = Data['Y2']

    Data_X1F = Data_X1 = Data_X1.astype(float) #Setting String Dataframe -> float.
    Data_Y1F = Data_Y1 = Data_Y1.astype(float) 
    Data_X2F = Data_X2 = Data_X2.astype(float) 
    Data_Y2F = Data_Y2 = Data_Y2.astype(float) 
    Data_X3F = Data_X3 = Data_X3.astype(float) 
    Data_Y3F = Data_Y3 = Data_Y3.astype(float) 

    Data_list_X1 = Data_X1F.values.tolist() #Setting float dataframe to list.
    Data_list_Y1 = Data_Y1F.values.tolist()
    Data_list_X2 = Data_X2F.values.tolist()
    Data_list_Y2 = Data_Y2F.values.tolist()
    Data_list_X3 = Data_X3F.values.tolist()
    Data_list_Y3 = Data_Y3F.values.tolist()
    
    
    #Calulating the Least square fit data line from raw data lists for a LOBF.
    #If data was in separate files, wouldnt have to specify legnth. range(len()) doesnt work. 
    Gradient1 = np.polyfit(Data_list_X1[0:231] , Data_list_Y1[0:231] , 1) #Finds [gradient, Y-intercept] using least squares fit via 1st degree poly
    print("Gradient, Si:", Gradient1) #Issue above, python thinks all lists = len(longest list) therefore "nan" for all blanks 
    Gradient2 = np.polyfit(Data_list_X2 , Data_list_Y2 , 1) #Finds [gradient, Y-intercept]
    print("Gradient, Film:", Gradient2)
    Gradient3 = np.polyfit(Data_list_X3[0:395] , Data_list_Y3[0:395] , 1) #Finds [gradient, Y-intercept]
    print("Gradient, Edge-to-Bulk:", Gradient3)
    
    #Difference in gradients between "Background" Data and interested inclined "Growth" data. 
    Gradient_ETB_Si = abs(Gradient3)-abs(Gradient1) #Diff in gradients between Si | film and Edge to Bulk (ETB)
    Gradient_ETB_Film = abs(Gradient3)-abs(Gradient2)
    print("Gradient, ETB-Si:", Gradient_ETB_Si)
    print("Gradient, ETB-Film:", Gradient_ETB_Film)
    
    #Generating Lines Of Best Fit (LOBF) from above LINEST (Least squares fit) gradient.
    ETB_Si_Y_DATA = []
    for i in range(len(Data_list_X3)): #m*x 
        ETB_Si_Y = Gradient_ETB_Si[0] * Data_list_X3[i]
        ETB_Si_Y_DATA.append(ETB_Si_Y)
    #print(ETB_Si_Y_DATA)
    ETB_Film_Y_DATA = []
    for i in range(len(Data_list_X3)): #m*x 
        ETB_Film_Y = Gradient_ETB_Film[0] * Data_list_X3[i]
        ETB_Film_Y_DATA.append(ETB_Film_Y)
    #print(ETB_Film_Y_DATA) 
    
    #Generating Final Gradient Shifted Data (GSD) with sloped removed. 
    GSD_ETB_Si = [] #GSD = Gradient shifted Data 
    for i in range(len(Data_list_Y3)):
        ETB_Si_DIFF = Data_list_Y3[i] - abs(ETB_Si_Y_DATA[i])
        GSD_ETB_Si.append(ETB_Si_DIFF)
    #print(GSD_ETB_Si)
    GSD_ETB_Film = [] #GSD = Gradient shifted Data 
    for i in range(len(Data_list_Y3)):
        ETB_Film_DIFF = Data_list_Y3[i] - abs(ETB_Film_Y_DATA[i])
        GSD_ETB_Film.append(ETB_Film_DIFF)
    #print(GSD_ETB_Film)
    
    def Graphing():
        plt.title("Gradient Shifted FTB by comparison of Si and Film data | BT15")
        plt.xlabel("Length (µm)")
        plt.ylabel("Height (nm)")
        #plt.ylim(0, 30000) 
        #plt.xlim(3450, 3700)
        plt.grid(True)
        plt.rcParams["figure.figsize"] = 10,6
        
        xD1 = Data_list_X3 #FTB RAW
        yD1 = Data_list_Y3
        plt.plot(xD1, yD1)
        xD2 = Data_list_X3 #FTB-Si
        yD2 = GSD_ETB_Si
        plt.plot(xD2, yD2)
        xD3 = Data_list_X3 #FTB-Film
        yD3 = GSD_ETB_Film
        plt.plot(xD3, yD3)
        plt.legend(["Film RAW", "Film-Si","Film-ETB"], loc ="upper left")
    Graphing() 
Gradient_3_Var()
```
